```python
import pandas as pd

# Read the data

df = pd.read_csv("numerai.csv")
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler, OrdinalEncoder

X = df.drop(columns=["target"])
Y = df[["target"]]

# Encode target
oe = OrdinalEncoder()
Y = oe.fit_transform(Y).ravel()

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
## Do not modify the train - test data split
## If you want to have a seperate part for validation, feel free to further split the training part (tx, ty).
## Do use the testing part (vx, vy) in validation

cut = int(len(df) * 0.8)

tx = X_scaled[:cut]
vx = X_scaled[cut:]
ty = Y[:cut]
vy = Y[cut:]

tx.shape, vx.shape, ty.shape, vy.shape
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, ExtraTreesClassifier

base_models = [
    ('rf', RandomForestClassifier(n_estimators=100, random_state=42)),
    ('gb', GradientBoostingClassifier(n_estimators=100, random_state=42)),
    ('et', ExtraTreesClassifier(n_estimators=100, random_state=42))
]

model = StackingClassifier(
    estimators=base_models,
    final_estimator=LogisticRegression(max_iter=1000)
)

model.fit(tx, ty)

# run for 48 mins

# training = 98.36%
# testing = 50.30%


```

```python
from sklearn.metrics import accuracy_score

predict = model.predict(tx)
print(f"Training accuracy = {(accuracy_score(predict, ty)):.2%}")

predict = model.predict(vx)
print(f"Testing accuracy = {(accuracy_score(predict, vy)):.2%}")
```
